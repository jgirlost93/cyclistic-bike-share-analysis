```python
## install and load libraries

import pandas as pd
import os
```

```python
## check file names

print("Files in current directory:")
print(os.listdir())
```

```python
## make sure filenames exactly match what appears in os.listdir()

q1_2019 = pd.read_csv("Divvy_Trips_2019_Q1.csv", low_memory=False)
q1_2020 = pd.read_csv("Divvy_Trips_2020_Q1.csv", low_memory=False)
```

```python
## ompare column names for each of the files

print("\nColumns for Q1 2019:\n", q1_2019.columns)
print("\nColumns for Q1 2020:\n", q1_2020.columns)
```

```python
## rename 2019 columns to match 2020 columns

q1_2019 = q1_2019.rename(columns={
    "trip_id": "ride_id",
    "bikeid": "rideable_type",
    "start_time": "started_at",
    "end_time": "ended_at",
    "from_station_name": "start_station_name",
    "from_station_id": "start_station_id",
    "to_station_name": "end_station_name",
    "to_station_id": "end_station_id",
    "usertype": "member_casual"
})
```

```python
## inspect dataframes
print("\nQ1 2019 Data Info:")
q1_2019.info()

print("\nQ1 2020 Data Info:")
q1_2020.info()
```

```python
## convert columns to string so they stack correctly

q1_2019["ride_id"] = q1_2019["ride_id"].astype(str)
q1_2019["rideable_type"] = q1_2019["rideable_type"].astype(str)

q1_2020["ride_id"] = q1_2020["ride_id"].astype(str)
q1_2020["rideable_type"] = q1_2020["rideable_type"].astype(str)
```

```python
## combine datasets

all_trips = pd.concat([q1_2019, q1_2020], ignore_index=True)
```

```python
## remove columns that were dropped beginning in 2020

all_trips = all_trips.drop(
    columns=["start_lat", "start_lng", "end_lat", "end_lng", "birthyear", "gender", "tripduration"],
    errors="ignore"
)
```

```python
## clean up and add data to prepare for analysis

print("\nList of column names:\n", all_trips.columns)
print("\nHow many rows are in data frame?", len(all_trips))
print("\nDimensions of the data frame:", all_trips.shape)
print("\nFirst 6 rows:\n", all_trips.head())
print("\nColumn info:\n")
all_trips.info()
print("\nStatistical summary:\n", all_trips.describe(include="all"))
```

```python
## clean member/casual values

print("\nValue counts for 'member_casual' before cleaning:\n", all_trips["member_casual"].value_counts(dropna=False))

all_trips["member_casual"] = all_trips["member_casual"].replace({
    "Subscriber": "member",
    "Customer": "casual"
})

print("\nValue counts for 'member_casual' after cleaning:\n", all_trips["member_casual"].value_counts(dropna=False))
```

```python
## convert date columns

all_trips["started_at"] = pd.to_datetime(all_trips["started_at"], errors="coerce")
all_trips["ended_at"] = pd.to_datetime(all_trips["ended_at"], errors="coerce")
```

```python
## add date-related columns

all_trips["date"] = all_trips["started_at"].dt.date
all_trips["month"] = all_trips["started_at"].dt.month
all_trips["day"] = all_trips["started_at"].dt.day
all_trips["year"] = all_trips["started_at"].dt.year
all_trips["day_of_week"] = all_trips["started_at"].dt.day_name()
```

```python
## add ride length in seconds

all_trips["ride_length"] = (all_trips["ended_at"] - all_trips["started_at"]).dt.total_seconds()

print("\nData types after adding columns:\n")
all_trips.info()
```

```python
## remove bad data

all_trips_v2 = all_trips[
    (all_trips["start_station_name"] != "HQ QR") &
    (all_trips["ride_length"] >= 0)
].copy()
```

```python
## conduct descriptive analysis

print("\nDescriptive statistics for ride_length:\n", all_trips_v2["ride_length"].describe())

print("\nRide length statistics grouped by member_casual:\n")
print(
    all_trips_v2.groupby("member_casual")["ride_length"].agg(["mean", "median", "max", "min"])
)

print("\nAverage ride length by member type and day of week:\n")
print(
    all_trips_v2.groupby(["member_casual", "day_of_week"])["ride_length"].mean()
)
```

```python
## order weekdays properly

days_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
all_trips_v2["day_of_week"] = pd.Categorical(
    all_trips_v2["day_of_week"],
    categories=days_order,
    ordered=True
)

print("\nAverage ride length (sorted by day of week):\n")
print(all_trips_v2.groupby(["member_casual", "day_of_week"], observed=True)["ride_length"].mean())
```

```python
## summary table

summary_stats = all_trips_v2.groupby(
    ["member_casual", "day_of_week"], observed=True
).agg(
    number_of_rides=("ride_id", "count"),
    average_duration=("ride_length", "mean")
).reset_index()

print("\nSummary of rides and duration by rider type and weekday:\n", summary_stats)
```

```python
## export summary file for future analysis

counts = all_trips_v2.groupby(
    ["member_casual", "day_of_week"], observed=True
)["ride_length"].mean().reset_index()

print("\nExport complete: avg_ride_length.csv")
```

```python
# Chart 1: Avg ride length by user type
avg_ride = all_trips_v2.groupby("member_casual", observed=True)["ride_length"].mean().reset_index()
avg_ride.to_csv("avg_ride_by_type.csv", index=False)

# Chart 2: Number of rides by day
rides_by_day = all_trips_v2.groupby(["member_casual", "day_of_week"], observed=True)["ride_id"].count().reset_index()
rides_by_day.to_csv("rides_by_day.csv", index=False)

# Chart 3: Avg ride length by day
avg_length_day = all_trips_v2.groupby(["member_casual", "day_of_week"], observed=True)["ride_length"].mean().reset_index()
avg_length_day.to_csv("avg_length_day.csv", index=False)
```
